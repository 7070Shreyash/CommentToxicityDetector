```python
import os
import pandas as pd
import tensorflow as tf
import numpy as np
```

```python
os.path.join('jigsaw-toxic-comment-classification-challenge','train.csv','train.csv')
```

```python
df = pd.read_csv(os.path.join('jigsaw-toxic-comment-classification-challenge','train.csv','train.csv')
)
```

```python
df.head()
```

```python
#Deep learning models only recog then conv the text into tokenization
```

```python
from tensorflow.keras.layers import TextVectorization
```

```python
x = df["comment_text"]
y = df[df.columns[2:]].values # for conv of the output into numpy array
```

```python
#Total no of unique words to be mapped by toknization must be predefined
```

```python
MAX_FEATURES = 200000 # no of words
```

```python
vectorizer = TextVectorization(max_tokens = MAX_FEATURES,output_sequence_length=1800,output_mode="int")
```

```python
vectorizer.adapt(x.values) #Mapping
```

```python
vectorizer("Hello there")
```

```python
x.values
```

```python
vectorized_text = vectorizer(x.values)
```

```python
#Using dataPipelnes
```

```python
dataset = tf.data.Dataset.from_tensor_slices((vectorized_text,y))
dataset = dataset.cache()
dataset = dataset.shuffle(160000)
dataset = dataset.batch(16) #16 dataset in one batch (x,y)
dataset = dataset.prefetch(8)
```

```python
batch_X, batch_y = dataset.as_numpy_iterator().next()
```

```python
train = dataset.take(int(len(dataset)*.7))
val = dataset.skip(int(len(dataset)*.7)).take(int(len(dataset)*.2))
test = dataset.skip(int(len(dataset)*.9)).take(int(len(dataset)*.1))
```

```python
#we alwasy train on the batches
```

```python
#Building the Deep Model
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout , Bidirectional, Dense,Embedding
```

```python
#Dropout for regularization , Embedding for knowing the cahr. of the words in each sense
```

```python
#32 info about a word in max_features
#Words prior to word and after also has a meaning but sequential allows only uni flow hece using bidirectional
#Ex- I dont hate you
#LSTM layer has optimization for tanh activation
#Now keras will do the thing according to the val things
#Final output has 6 diff outputs and ind is either 0 or 1 using sigmoid
```

```python
model = Sequential()
model.add(Embedding(MAX_FEATURES+1,32));

model.add(Bidirectional(LSTM(32,activation = "tanh")))

model.add(Dense(128,activation = "relu"))

model.add(Dense(256,activation = "relu"))

model.add(Dense(128,activation = "relu"))

model.add(Dense(6,activation = "sigmoid"))

```

```python
model.compile("adam",loss = tf.losses.BinaryCrossentropy())
#Because values are 6 diff binary digits , it helps in backP
```

```python
model.summary()
```

```python
logdir = "logs"
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir = logdir)
```

```python
hist = model.fit(train,epochs = 1,validation_data = val,callbacks = [tensorboard_callback])
```

```python
#Evaluation of the Models
```

```python
from tensorflow.keras.metrics import Precision , Recall , CategoricalAccuracy
```

```python
#These are appended layers of the data and can bu updated as long as we will get the next set of the batches
```

```python
pre = Precision()
re = Recall()
acc = CategoricalAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    
    X_true , Y_true = batch
    Y_hat = model.predict(X_true)
    
    #conv into one long array
    Y_true = Y_true.flatten()
    Y_hat = Y_hat.flatten()
    
    pre.update_state(Y_true,Y_hat)
    re.update_state(Y_true,Y_hat)
    acc.update_state(Y_true,Y_hat)
    
```

```python
print(f'Precision : {pre.result().numpy()} , Recall : {re.result().numpy()} , Accuracy : {acc.result().numpy()}')
```
